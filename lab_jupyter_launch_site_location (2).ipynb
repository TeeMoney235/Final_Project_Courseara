```python
# First, install and import required packages
import piplite
await piplite.install(['folium'])
await piplite.install(['pandas'])

import folium
import pandas as pd
from folium.plugins import MarkerCluster
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

# Distance calculation function
def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points
    on the Earth's surface using the Haversine formula
    
    Parameters:
    lat1, lon1: Latitude and Longitude of point 1 (in decimal degrees)
    lat2, lon2: Latitude and Longitude of point 2 (in decimal degrees)
    
    Returns:
    Distance in kilometers
    """
    # Approximate radius of Earth in km
    R = 6373.0
    
    # Convert coordinates from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    # Calculate differences
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    
    return distance

# Download and load the SpaceX dataset
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Get launch site coordinates (using CCAFS SLC-40 as example)
launch_site = spacex_df[spacex_df['Launch Site'] == 'CCAFS SLC-40'].iloc[0]
launch_site_lat = launch_site['Lat']
launch_site_lon = launch_site['Long']

# Define coordinates of nearby features (example values - you should verify these)
coastline_lat, coastline_lon = 28.56367, -80.57163  # Approximate coastline point
highway_lat, highway_lon = 28.56353, -80.57085      # Nearby highway
railway_lat, railway_lon = 28.57206, -80.58528      # Nearby railway
city_lat, city_lon = 28.10469, -80.64531            # Titusville city center

# Calculate distances
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

# Print results
print(f"Distance to coastline: {distance_coastline:.2f} km")
print(f"Distance to highway: {distance_highway:.2f} km")
print(f"Distance to railway: {distance_railway:.2f} km")
print(f"Distance to city (Titusville): {distance_city:.2f} km")

# Create a map visualization
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="CCAFS SLC-40",
    icon=folium.Icon(color='red')
).add_to(site_map)

# Add proximity markers and lines
proximities = [
    (coastline_lat, coastline_lon, distance_coastline, "Coastline", "blue"),
    (highway_lat, highway_lon, distance_highway, "Highway", "green"),
    (railway_lat, railway_lon, distance_railway, "Railway", "orange"),
    (city_lat, city_lon, distance_city, "Titusville", "purple")
]

for lat, lon, distance, name, color in proximities:
    # Add marker
    folium.Marker(
        [lat, lon],
        popup=name,
        icon=folium.Icon(color=color)
    ).add_to(site_map)
    
    # Add line
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [lat, lon]],
        color=color,
        weight=2
    ).add_to(site_map)

# Display the map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Import required libraries
import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

# Define distance calculation function
def calculate_distance(lat1, lon1, lat2, lon2):
    """Calculate distance between two points using Haversine formula"""
    R = 6373.0  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return R * 2 * atan2(sqrt(a), sqrt(1-a))

# CCAFS SLC-40 coordinates
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Proximity points coordinates (found using MousePosition)
proximities = {
    'Coastline': {'lat': 28.56319, 'lon': -80.56799, 'color': 'blue'},
    'Highway': {'lat': 28.56385, 'lon': -80.57085, 'color': 'green'}, 
    'Railway': {'lat': 28.57206, 'lon': -80.58528, 'color': 'orange'},
    'City (Titusville)': {'lat': 28.61222, 'lon': -80.80788, 'color': 'purple'}
}

# Create map centered on launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    icon=folium.Icon(color='red'),
    popup='CCAFS SLC-40 Launch Site'
).add_to(site_map)

# Add markers and lines for each proximity point
for name, data in proximities.items():
    # Calculate distance
    distance = calculate_distance(launch_site_lat, launch_site_lon, 
                                data['lat'], data['lon'])
    
    # Create marker with distance label
    folium.Marker(
        [data['lat'], data['lon']],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color: {data["color"]};"><b>{name}: {distance:.2f} KM</b></div>'
        ),
        popup=name
    ).add_to(site_map)
    
    # Draw connecting line
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [data['lat'], data['lon']]],
        color=data['color'],
        weight=2
    ).add_to(site_map)

# Add MousePosition for coordinate finding
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Display the map
site_map
```
